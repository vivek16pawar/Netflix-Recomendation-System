## Data Cleaning

```python
df.columns
```

```python
df = df[['movie_id','title','overview','cast','crew','keywords','genres']]
```

```python
df.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text) :
       L.append(i['name'])
    return L
```

```python
df.head()
```

```python
df.genres[0]
```

```python
df.genres = df.genres.apply(convert)
```

```python
df.head()
```

```python
df.keywords[0]
```

```python
df.keywords = df.keywords.apply(convert)
```

```python
df.head()
```

```python
def convert2(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text) :
        if counter < 5:
             L.append(i['name'])
        counter += 1
    return L
```

```python
df.cast = df.cast.apply(convert2)
```

```python
df.head()
```

```python
df.cast[0]
```

```python
df.crew[1]
```

```python
def director(text):
    D = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            D.append(i['name'])
    return D
```

```python
df.crew = df.crew.apply(director)
```

```python
df.head()
```

```python
def joining(text):
    J = []
    for i in text:
        J.append(i.replace(' ',''))
    return J
```

```python
df.cast = df.cast.apply(joining)
df.crew = df.crew.apply(joining)
df.keywords = df.keywords.apply(joining)
df.genres = df.genres.apply(joining)
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
df['tags'] = df['overview'] + df['cast'] + df['crew'] + df['keywords'] + df['genres']
```

```python
df.head()
```

```python
df.tags[0]
```

```python
new_df = df.drop(columns = ['cast','crew','overview','keywords','genres'])
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x))
```

## Model Building

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
new_df[new_df['title'] == 'Batman'].index[0]
```

```python
sorted(similarity[1362],reverse = True)[0:5]
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommendation(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
   
```

```python
recommendation('Batman')
```
